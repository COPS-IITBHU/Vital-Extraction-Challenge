```python
import os
os.chdir('/content/drive/MyDrive/Classification Dataset/')
import pandas as pd
import numpy as np
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
import cv2
img_classes = ['BPL-EliteView-EV100-C', 'BPL-Ultima-PrimeD-A', 'Nihon-Kohden-lifescope-A', 'BPL-EliteView-EV10-B_Meditec-England-A']
attrs = ['SPO2', 'HR', 'MAP']

thresholds = {
    'SPO2' : 200,
    'HR': 200,
    'MAP': 150
}

averages = {}
areas = {}
pts = {}
w_s = {}
h_s = {}
green_mask = []
for attr in attrs:
  averages[attr] = []
  pts[attr] = []
  areas[attr] = []
  w_s[attr] = []
  h_s[attr] = []

labels = []
X = []
yellow_mask = []
magenta_mask = []

for j in [2]:
  img_class = img_classes[j]
  df = pd.read_csv(f'{img_class}.csv')
  for index in range(len(df)):
    img_name = df.iloc[index]['image_name']
    
    img = cv2.imread(f'./{img_class}/{img_name}')
    try: 
      arr = []
      for attr in attrs:
        x, y, w, h = eval(df.iloc[index][attr])
        x *= img.shape[1]
        w *= img.shape[1]
        y *= img.shape[0]
        h *= img.shape[0]
        areas[attr].append(h*w)
        cropped = img[int(y-h/2):int(y+h/2), int(x-w/2):int(x+w/2)]
        #_, cropped = cv2.threshold(cropped,thresholds[attr],255,cv2.THRESH_BINARY)
        #plt.imshow(cropped)
        #plt.show()
        pts[attr].append([x, y])
        average = cropped.mean(axis=0).mean(axis=0)
        averages[attr].append(average)
        h_s[attr].append([h])
        w_s[attr].append([w])
        X.append([*average, x, y, h*w])
        labels.append([attr])
        if attr == 'HR':
          hsv = cv2.cvtColor(cropped, cv2.COLOR_BGR2HSV)
          mask = cv2.inRange(hsv, (60, 60, 60), (70, 255,255))
          imask = mask>0
          green_mask.append(imask.sum()/(w*h))
        if attr == 'SPO2':
          hsv = cv2.cvtColor(cropped, cv2.COLOR_BGR2HSV)
          mask = cv2.inRange(hsv, (28, 90, 100), (40, 255, 255))
          imask = mask>0
          yellow_mask.append(imask.sum()/(w*h))
        if attr == 'MAP':
          hsv = cv2.cvtColor(cropped, cv2.COLOR_BGR2HSV)
          mask = cv2.inRange(hsv, (122, 20, 70), (155, 255, 255))
          imask = mask>0
          green = np.zeros_like(cropped, np.uint8)
          green[imask] = cropped[imask]
          magenta_mask.append(imask.sum()/(w*h))
    except Exception as e:
      print(e)
      continue
  
```

```python
import torch
from craft_text_detector import (
    read_image,
    load_craftnet_model,
    load_refinenet_model,
    get_prediction,
    export_detected_regions,
    export_extra_results,
    empty_cuda_cache
)
from transformers import TrOCRProcessor, VisionEncoderDecoderModel
processor = TrOCRProcessor.from_pretrained('microsoft/trocr-large-printed')
model = VisionEncoderDecoderModel.from_pretrained('microsoft/trocr-large-printed')

DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
model = model.to(DEVICE)

def get_boxes(image_path):
  image = read_image(image_path)
  refine_net = load_refinenet_model(cuda=True)
  craft_net = load_craftnet_model(cuda=True)
  prediction_result = get_prediction(
    image=image,
    craft_net=craft_net,
    refine_net=refine_net,
    text_threshold=0.7,
    link_threshold=0.4,
    low_text=0.4,
    cuda=True,
    long_size=1280
  )
  empty_cuda_cache()
  return prediction_result['boxes']
```
